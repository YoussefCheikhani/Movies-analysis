```python
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)         #Adjust configuration of the plot

#Read from data

df = pd.read_csv(r'C:\Users\youss\Desktop\Career\Datasets\movies.csv')
```

```python
#Display data
df.head()
```

```python
#Check for missing data (null values)

for col in df.columns:
    pct_missing = str(np.mean(df[col].isnull())*100)+'%'
    print(col,":",str(pct_missing))            #Budget column has the highest  missing values
    
    
```

```python
#for simplicity purpose we will delete rows with missing values 
df = df.dropna()
```

```python
#Columns Data types Check
df.dtypes
```

```python
#Change data types for columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#We noticed that year column is not accurate in certain rows
#Create yearcorrect column from released using regular expression

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int) 
```

```python
df.head()
```

```python
#Sorting data by gross
df.sort_values(by=['gross'],inplace=False, ascending=False).head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Drop duplicates

df.drop_duplicates().head()
```

```python
#Finding high correlations variables to gross revenues for movies
```

```python
#Scatter plot for budget vs gross

#plt.ylim(0, 990000000)         #Gross range rounding (uncomment to scale gross by 100 millions)
##plt.xlim(3500000,19000000)            #Budget Range

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross revenue')

plt.xlabel('Budget')

plt.ylabel('Gross revenue')

plt.show()
```

```python
# reg plot using seaborn

sns.regplot(data=df,x='budget',y='gross', scatter_kws={"color":"red"}, line_kws={"color":"yellow"}) 
```

```python
#Correlations Matrix
df.corr()       #Pearson by default (try also Kendall, Spearman)
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Values")

plt.show()
```

```python
# Two strong positive correlations budget/gross & votes/gross
```

```python
#Finding correlation for Non Numeric columns

df_numerized = df

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')      #change type of column to category
        df_numerized[col_name] = df_numerized[col_name].cat.codes               #asign random number to each category in the column    
    
df_numerized.head()                                                        
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Values")

plt.show()
```

```python
df_numerized.corr()
```

```python
#Sorting correlations
cor_mat = df_numerized.corr()

cor_pairs = cor_mat.unstack()

sorted_cor_pairs = cor_pairs.sort_values(ascending=False)

strong_correlation = sorted_cor_pairs[(sorted_cor_pairs)>0.5]
strong_correlation
```
